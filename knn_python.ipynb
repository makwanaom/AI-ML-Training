```python
sf_index = np.random.permutation(df.shape[0])
```

```python
data  = df.iloc[sf_index].values
data
```

```python
def Euclidean_Distance(Point1, Point2):
    if len(Point1) == len(Point2):
        return np.sqrt(np.sum((Point1[:4] - Point2[:4])**2))
    else:
        print("Shape should be same..")
```

```python
Euclidean_Distance(data[0], data[0])
```

```python
def Get_NN(TrainingData, testRow, K=3):
    Distance = []
    for oneRow in TrainingData:
        dis = Euclidean_Distance(oneRow, testRow)
        Distance.append(dis)
    

    Distance  = np.array(Distance)
    TrainingData = np.array(TrainingData)
    dis_index = Distance.argsort()
    TrainingData = TrainingData[dis_index]
    Nn = TrainingData[:K]
    Classes = []
    for n in Nn:
        Classes.append(n[-1])
    predClass = max(Classes, key= Classes.count)
    return predClass


Get_NN(data, data[10])



def Predict(testData):
    trainData = data
    out = []
    for test in testData:
        pred = Get_NN(trainData, test)
        out.append(pred)
    return np.array(out)


pred  = Predict(data)
pred 
```
